```python
def get_model(in_particles=150, num_features=3, ff_dim=64, dropout=0.0, numheads=4):
    """Create a transformer model."""
    ###################################################################### below is layer 1
    input_layer = keras.layers.Input(shape=(in_particles, num_features), name='input')

    norm = keras.layers.BatchNormalization(name='norm')(input_layer)

    emb = keras.layers.Dense(ff_dim, activation=keras.activations.relu)(norm)

    x = keras.layers.MultiHeadAttention(numheads, ff_dim//numheads,
                                        name='mha1', dropout=dropout)(emb, emb, emb)

    res = keras.layers.Add()([x, emb])
    ###################################################################### below is layer 2
    x = keras.layers.Dense(ff_dim, activation=keras.activations.relu)(res)
    x = keras.layers.Dense(ff_dim, activation=keras.activations.relu)(x)

    lo1 = keras.layers.Add()([x, res])

    x = keras.layers.MultiHeadAttention(numheads, ff_dim//numheads,
                                        name='mha2', dropout=dropout)(lo1, lo1, lo1)

    res = keras.layers.Add()([x, lo1])
    
    ####################################################################### below is layer 3
    x = keras.layers.Dense(ff_dim, activation=keras.activations.relu)(res)
    x = keras.layers.Dense(ff_dim, activation=keras.activations.relu)(x)
    lo2 = keras.layers.Add()([x, res])
    ####################################################################### below is layer 4 - out
    x = keras.layers.GlobalAveragePooling1D(data_format='channels_last')(lo2)
    x = keras.layers.Dense(ff_dim, activation="relu")(x)
    out = keras.layers.Dense(5)(x)

    model = keras.Model(input_layer, out)

    model.summary()
    return model
```
